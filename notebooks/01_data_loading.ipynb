```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/raw/India_Flood_Inventory_v3.csv")

# Preview first 5 rows
df.head()
```

```python
# Check the shape of the data (rows, columns)
print("Shape:", df.shape)

# View column names
print("\nColumns:", df.columns.tolist())

# View data types and nulls
df.info()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
df.columns
```

```python
df["start_date"] = pd.to_datetime(df["start_date"], errors="coerce")
df["end_date"] = pd.to_datetime(df["end_date"], errors="coerce")
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df = df.dropna(subset=["state", "start_date"])
```

```python
df = df.drop(columns=["unnamed:_0"], errors="ignore")
```

```python
df.to_csv("../data/clean/flood_cleaned.csv", index=False)
```
